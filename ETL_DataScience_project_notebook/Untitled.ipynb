```python
import pandas as pd
#import psycopg2 as ps
from sqlalchemy import create_engine
```

```python
csv_file_all_data = "../Resources/fulltimeAL.csv"
DS_jobs_all_data_df = pd.read_csv(csv_file_all_data)
DS_jobs_all_data_df.head()
```

```python
#Create new data with selected columns- position, company, description, location
new_DS_jobs_all_data_df = DS_jobs_all_data_df[["position", "company","description","location"]].copy()
new_DS_jobs_all_data_df.head(6)
```

```python
#Transformation Step: Split location into City, "State Zip"
#CityStateZip_df = new_DS_jobs_all_data_df["location"].str.extract('(?P<City>[A-Z ]{1,2}),(?P<State>[A-Z]*$) (?P<Zipcode>\d{5}){1,1}', expand=True)                                     
#CityStateZip_df = new_DS_jobs_all_data_df["location"].str.extract('(?P<City>([A-Z ]\S+){2,})', expand=True)
new_DS_jobs_all_data_df['City'], new_DS_jobs_all_data_df['State']  = new_DS_jobs_all_data_df["location"].str.split(', ',1).str
new_DS_jobs_all_data_df['Zipcode'] = new_DS_jobs_all_data_df['State'].str[2:]
new_DS_jobs_all_data_df['State'] = new_DS_jobs_all_data_df['State'].str[:2]
new_DS_jobs_all_data_df.head()
```

```python
new_DS_jobs_all_data_df.drop(columns=["location"], inplace=True)
new_DS_jobs_all_data_df.head()
```
